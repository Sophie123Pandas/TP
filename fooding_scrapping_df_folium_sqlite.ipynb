```python
import json
import requests
import pandas as pd
```

```python
#modification de l'url pour pouvoir scrapper toutes les pages
url = 'https://lefooding.com/fr/search?type=restaurant&page=1&sort=&per_page=10&total_pages=98&total_entries=972'
```

```python
resp = requests.get(url=url)
print(url, resp.status_code)
d = resp.json()
```

```python
name=[]
typ=[]
city=[]
country=[]
cp=[]
addr=[]
lat=[]
long=[]
phone=[]
prix=[]
spe1=[]
spe2 = []
spe3 = []

for i in range(len(d['results'])):
    name.append(d['results'][i]['name'])
    typ.append(d['results'][i]['type'])
    city.append(d['results'][i]['address']['city'])
    country.append(d['results'][i]['address']['country'])
    cp.append(d['results'][i]['address']['postal_code'])
    addr.append(d['results'][i]['address']['street_name_1'])
    try: 
        lat.append(d['results'][i]['address']['position'][0])
    except: 
        lat.append('')
    try: 
        long.append(d['results'][i]['address']['position'][1])
    except: 
        long.append('')
    phone.append(d['results'][i]['phone_num'])
    prix.append(d['results'][i]['price'])
    try:
        spe1.append(d['results'][i]['cravings'][0])
    except:
        spe1.append('') 
    try:
        spe2.append(d['results'][i]['cravings'][1])
    except:
        spe2.append("none")
    try:
        spe3.append(d['results'][i]['cravings'][2])
    except:
        spe3.append("none")
```

```python
foodfood = pd.DataFrame(
    list(zip(name, typ, city, country, cp, addr, lat, long, phone, prix, spe1, spe2, spe3)),
    columns =['name', 'type', "city", "country", "cp", "addr", "lat", "long", "phone", "prix", "spe1", "spe2", "spe3"])
```

```python
foodfood
```

```python
foodfood['name']= foodfood['name'].astype(str)
foodfood['type']= foodfood['type'].astype(str)
foodfood["city"]= foodfood['city'].astype(str)
foodfood["cp"]= foodfood['cp'].astype(str)
foodfood["addr"]= foodfood['addr'].astype(str)
foodfood["lat"]= pd.to_numeric(foodfood['lat'])
foodfood["long"]= pd.to_numeric(foodfood['long'])
foodfood["phone"]= foodfood['phone'].astype(str)
foodfood["prix"]= foodfood['prix'].astype(str)
foodfood["spe1"]= foodfood['spe1'].astype(str)
foodfood["spe2"]= foodfood['spe2'].astype(str)
foodfood["spe3"]= foodfood['spe3'].astype(str)
foodfood.info()
```

```python
foodfood.lat.isnull().sum().sum()
```

```python
foodfood = foodfood[foodfood.name != 'Claire et Hugo']
```

```python
foodfood.to_csv('fooding.csv')
```

```python
import folium
from folium.plugins import MarkerCluster
```

```python
df = foodfood
rest_map = folium.Map(location=[48.8,2.35],zoom_start=10, tiles='OpenStreetMap')

marker_cluster = folium.plugins.MarkerCluster().add_to(rest_map)

zone='75020'
cuisine="Asiatique"


for i in df[(df['cp']==zone) & ((df['spe1']==cuisine) | (df['spe2']==cuisine) | (df['spe3']==cuisine))].index:
    marker = folium.Marker(location=(round(df['lat'].loc[i],6),round(df['long'].loc[i],6)), popup=(str(df['name'].loc[i]) +' ||| '+'Telephone'+str(df['phone'].loc[i],))).add_to(marker_cluster)

rest_map
```

Creation base SQLite

```python
import sqlite3
```

```python
conn = sqlite3.connect('DB.db')
c = conn.cursor()
c.execute('CREATE TABLE DBFOODING (name, type, city, cp, addr, lat, long, phone, prix, spe1, spe2, spe3)')
conn.commit()
foodfood.to_sql('DBFOODING', conn, if_exists='replace', index = False)
```
